### Word Imbedding

```python
'''
OHE로 만들어진 벡터는 sparse하고 대부분 0, 수동 인코딩
Word Imbedding은 그에 반해 밀집된 저차원 실수 벡터, 데이터로 부터 학습
'''
```

```python
# Embedding Layer 객체 생성
from keras.layers import Embedding
embedding_layer = Embedding(1000, 64) # 임베딩 layer는 적어도 2개의 parameter / 토큰갯수, 차원
```

```python
# IMDB data load
from keras.datasets import imdb
from keras import preprocessing

max_features = 10000 # 특성 단어 수
maxlen = 20

(x_train,y_train), (x_test,y_test) = imdb.load_data(num_words=max_features)

x_train = preprocessing.sequence.pad_sequences(x_train, maxlen)
x_test= preprocessing.sequence.pad_sequences(x_test, maxlen)
```

```python
from keras.models import Sequential
from keras.layers import Flatten, Dense, Embedding

model = Sequential()
model.add(Embedding(10000,8,input_length = maxlen)) # input/ 출력은 (sample,maxlen,8)
model.add(Flatten())
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['acc'])
model.summary()
```

```python
history = model.fit(x_train,y_train,
                 epochs=200,
                 batch_size=32,
                 validation_split=0.2)
```
