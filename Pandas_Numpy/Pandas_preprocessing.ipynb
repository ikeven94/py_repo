```python
# 결측값 처리(대체) : df.fillna()
df= pd.DataFrame(np.random.randn(5,3),
            columns=['c1','c2','c3'])
df
```

```python
df.loc[1,['c1','c3']]=None
df
```

```python
# np.where 이용하여 같은 결과
# np.where(조건, 참, 거짓)
```

```python
np.where(df.isnull(),df.mean(),df)
```

```python
pd.DataFrame(np.where(df.isnull(),df.mean(),df))
```

```python
df.where(pd.notnull(df), df.mean(), axis='columns') # column 단위로 수행
```

```python
df.where(pd.notnull(df),df.mean(),axis='columns',inplace=True)
df
# 데이터프레임, 시리즈 모두 where 함수 사용법 동일
```

```python
df2 = pd.DataFrame({'c1':[1,2,3,4,5],
             'c2':[6,7,8,9,10]})
df2
```

```python
# 7, 9 값을 np.nan으로 변경
df2.loc[[1,3],['c2']] = np.nan
```

```python
df2
# c2 new 열을 추가 (c2열 값이 null이 아니면 c2, 그렇지 안으면 c1열 값으로 채움)
df2['c2new'] = np.where(pd.notnull(df2['c2']),df2['c2'],df2['c1'])
df2
```

```python
for i in df2.index:
    if pd.notnull(df2.loc[i,'c2']) == True: # null이 아닌경우
        df2.loc[i,'c2new2'] = df2.loc[i,'c2']
    else:
        df2.loc[i,'c2new2'] = df2.loc[i,'c1']
df2
```
