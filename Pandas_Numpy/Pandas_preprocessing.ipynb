```python
df1 = pd.read_csv('C:/Users/USER/Desktop/dataset/deepL/data_merge_exdata/concat_1.csv')
df2 = pd.read_csv('C:/Users/USER/Desktop/dataset/deepL/data_merge_exdata/concat_2.csv')
df3 = pd.read_csv('C:/Users/USER/Desktop/dataset/deepL/data_merge_exdata/concat_3.csv')
```

```python
newRowSeries = pd.Series(['n1','n2','n3','n4'])
```

```python
pd.concat([df1,newRowSeries])
# 새로운 행 추가? 아님 >> 새로운 열로 추가됨
# Series는 열 이름이 존재하지 않으므로 새로운 행으로 연결 불가
# 다만 '0' 이라는 이름의 column으로 추가
```

```python
# Series를 dataframe 행으로 추가하고 싶다면
newRowdf = pd.DataFrame([['n1','n2','n3','n4']], columns=['A','B','C','D'])
```

```python
pd.concat([df1,newRowdf])
# concat은 2개 이상의 데이터 프레임 연결
```

```python
df1.append(newRowdf)
# append는 1개의 dataframe
```

```python
dataDict = {'A':'n1', 'B':'n2', 'C':'n3', 'D':'n4'}
dataDict
```

```python
df1
pd.concat([df1,dataDict]) # Error >> ignore_index를 주지 않아서
```

```python
df1.append(dataDict,ignore_index=True) # append 함수, 데이터 프레임과 딕셔너리 연결
```

```python
rowConcati = pd.concat([df1,df2,df3],ignore_index=True)
rowConcati
```

```python
colConcat = pd.concat([df1,df2,df3],axis=1) # 열 방향으로 concat
colConcat['A'] # A열만 추출
```

```python
# 우측에 newColumnLIst 칼럼 추가, Column data : n1,n2,n3,n4
colConcat['newColumlist'] = ['n1','n2','n3','n4']
colConcat
```

```python
pd.concat([df1,df2,df3],axis=1,ignore_index=True)  # 좌우 연결, 열이름도 새로
```

```python
df2.columns = ['E','F','G','H']
df2
```

```python
df3.columns = ['A','C','F','H']
df3
```

```python
# 중복된것도, 다른것도 있는 df
pd.concat([df1,df2,df3]) # 3개 df의 모든 열에대한 데이터 다 나옴
```

```python
# 공통열만 선택해서 연결 'inner'
print(pd.concat([df1,df2,df3],join = 'inner')) # 세 데이터의 공통 열은 존재하지 않으므로 empty
```

```python
# df1, df3 는 A, C 공통 열 존재
pd.concat([df1,df3],join='inner')
```

```python
df2.index = [4,5,6,7] #indexing
df2
```

```python
df3.index = [0,2,5,7]
df3
```

```python
pd.concat([df1,df2,df3]) # 위 아래 연결
```

```python
pd.concat([df1,df2,df3],axis=1) # 좌우 연결
```

```python
print(pd.concat([df1,df3], axis=1, join='inner')) #0 과 2라는 공통 열
```
