```python
# Merge 활용
person = pd.read_csv('C:/Users/USER/Desktop/dataset/deepL/data_merge_exdata/survey_person.csv')
site = pd.read_csv('C:/Users/USER/Desktop/dataset/deepL/data_merge_exdata/survey_site.csv')
survey = pd.read_csv('C:/Users/USER/Desktop/dataset/deepL/data_merge_exdata/survey_survey.csv')
visited = pd.read_csv('C:/Users/USER/Desktop/dataset/deepL/data_merge_exdata/survey_visited.csv')
```

```python
visited_subset = visited.loc[[0,2,6],]
visited_subset
```

```python
site
```

```python
pd.concat([visited_subset,site]) # 위아래
```

```python
pd.concat([visited_subset,site],axis=1) # 좌우
```

```python
# site의 name과 visited_subset의 site열 기준으로 merge
site.merge(visited_subset, left_on='name', right_on='site')
```

```python
# site와 visited 연결
site.merge(visited, left_on='name', right_on='site' )
```

```python
# person 과 survey 연결
person.merge(survey, left_on='ident', right_on='person')
```

```python
ps = person.merge(survey,left_on='ident',right_on='person')
vs = visited.merge(survey,left_on='ident',right_on='taken')
```

```python
pvs = ps.merge(vs, left_on=['ident','taken','quant','reading'],
               right_on = ['person','ident','quant','reading'])
pvs
```

```python
df = pd.DataFrame(np.arange(10).reshape(5,2),
                 index=['a','b','c','d','e'],
                 columns=['C1','C2'])
df
```

```python
df.sum() # 열 단위 덧셈, Nan은 0으로 취급
```
